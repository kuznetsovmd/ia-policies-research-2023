```python
import __path__
```

```python
from config import MAIN_EXPERIMENT


H1, H2, H3, H4, H5 = MAIN_EXPERIMENT.nb_args[product]
DESCRIPTOR = MAIN_EXPERIMENT.dataset.descriptor
PICTURES = MAIN_EXPERIMENT.pictures
```

```python
import json

from src.utils import resolve_stats


LAYOUT = dict(
    font=dict(family='Times New Roman', size=20, color='black'),
    plot_bgcolor='rgba(0, 0, 0, 0)',
    margin=dict(pad=10),
    height=400,
    width=900)

AXIS = dict(
    zerolinecolor='lightgrey',
    gridcolor='lightgrey',
    tickformat='000',
    showgrid=True, 
    tickangle=45, 
    ticklen=10)


hashes = []
unique_stats = []

with open(DESCRIPTOR, 'r', encoding='utf-8') as f:
    stats = json.load(f)

for s in stats:
    if s['policy_hash'] not in hashes and s['statistics'] is not None:
        hashes.append(s['policy_hash'])
        unique_stats.append(resolve_stats(s['statistics']))

from pprint import pprint
pprint(unique_stats)
```

```python
import plotly.graph_objects as go

from src.utils import make_hist3
from config import COLORS_D


bars, min_, step = make_hist3([stat['headings'] for stat in unique_stats], *H1)

fig = go.Figure()

fig.add_bar(
    x=[*[f'≥ {(i * step) + 1}' for i in range(len(bars) - 1)], f'≥ {(len(bars) - 1) * step + 1}'], y=bars,
    showlegend=False)

fig.update_layout(
    **LAYOUT,
    colorway=[COLORS_D[4]])

fig.update_xaxes(
    title=dict(
        font=dict(family='Times New Roman', size=25, color='black'),
        text='Количество заголовков'), 
    **AXIS)

fig.update_yaxes(
    title=dict(
        font=dict(family='Times New Roman', size=25, color='black'),
        text='Количество документов'), 
    **AXIS)

fig
```

```python
fig.write_image(f'{PICTURES}/headings.png')
```
